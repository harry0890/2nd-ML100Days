```python
print("(row總數, col總數)",app_train.shape)
print("dataset敘述: ")
app_train.describe()
```

```python
app_train['AMT_ANNUITY'].hist(grid=False, bins=15)
```

```python
print("最大值:%.2f" %app_train['AMT_ANNUITY'].max())
print("平均值:%.2f" %app_train['AMT_ANNUITY'].mean())
print("中位數:%.2f" %app_train['AMT_ANNUITY'].median())
print("標準差:%.2f" %app_train['AMT_ANNUITY'].std())
```

```python
print("最大值:%.2f" %app_train['AMT_INCOME_TOTAL'].max())
print("平均值:%.2f" %app_train['AMT_INCOME_TOTAL'].mean())
print("中位數:%.2f" %app_train['AMT_INCOME_TOTAL'].median())
print("標準差:%.2f" %app_train['AMT_INCOME_TOTAL'].std())
```

```python
plt.figure(figsize = (15,10))

Annuity = plt.subplot(2,1,1)
Annuity = app_train['AMT_ANNUITY'].hist(bins = 'auto', range = (0, 45000), color = 'skyblue')
Annuity.set_title('Distribution of AMT Annuity')
Annuity.set_xlabel('Amount of AMT annuity')
Annuity.set_ylabel('Frequency')

INCOME = plt.subplot(2,1,2)
INCOME = app_train['AMT_INCOME_TOTAL'].hist(bins = 'auto', range = (0,500000), color = 'red')
INCOME.set_title('Distribution of AMT INCOME')
INCOME.set_xlabel('Amount of AMT INCOME')
INCOME.set_ylabel('Frequency')
```
